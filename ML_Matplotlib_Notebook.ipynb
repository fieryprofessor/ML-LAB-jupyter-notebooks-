```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10,100)
y = np.sin(x)

plt.plot(x,y)
plt.title('Simple Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
x = np.linspace(-150,150,1000)
y = x**2
plt.plot(x,y)
plt.title('Parabola')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
date = ['25/12','26/12','27/12']
temp = [8.5,10.5,6.8]
plt.plot(date,temp)
plt.xlabel("Date") #add label to x-axis
plt.ylabel("Temperature") #add label to y-axis
plt.title("Data wise Temperature") #add the title to the chart
plt.grid(True) #add gridlines to the chart
plt.yticks(temp)
plt.show()
```

```python
y1 = np.cos(x)
plt.plot(x,y,label='sin(x)')
plt.plot(x,y1,label='cos(x)')
plt.title('Multiple Lines')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
fig, axs= plt.subplots(2)
axs[0].plot(x,y,'y')
axs[0].set_title('sin(x)')
axs[1].plot(x,y1,'g')
axs[1].set_title('cos(x)')
plt.tight_layout()
plt.show()
```

```python
x_vals = ['A','B','C','D','E']
y_vals = [3,7,5,4,6]
plt.bar(x_vals,y_vals,color='cyan',label='Bar')
plt.title('Bar Plot Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
data = np.random.randn(1000)
plt.hist(data,bins=30,color='purple',alpha=0.7)
plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
#Histogram using Numpy
import numpy as np
import matplotlib.pyplot as plt
x1 = np.random.normal(0,1,1000) #normal(mean,std,size)
x2 = np.random.normal(-2,1,1000)
x3 = np.random.normal(3,2,1000)
plt.hist(x1,bins=10,color='red',edgecolor='black')
plt.hist(x2,bins=10,color='green',edgecolor='black')
plt.hist(x3,bins=10,color='blue',edgecolor='black')
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 1000*np.random.rand(50)

plt.scatter(x,y,c=colors,s=sizes,alpha=0.5)
plt.title('Scatter Plot Example')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
np.random.seed(42) #For reproducibility
X = np.random.rand(50)
Y = np.random.rand(50)
plt.scatter(X,Y,color='red',marker='o',s=100,edgecolors='black',alpha=0.7,label='Scatter Plot')
#s-> sets the marker size
#edgecolors -> adjusts the transparency of the markers
plt.legend(loc=9)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(10)
d = np.random.normal(100,20,200)
fig = plt.figure(figsize=(10,7))
plt.boxplot(d)
plt.show()
```
