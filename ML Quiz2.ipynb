```python
#Create Dataframe with duplicate rows:
#a. Identify the duplicates and display only the duplicate row.
#b. drop duplicates while
# keeping the last occurence
#removing all duplicates
import pandas as pd
d={
    'Name':['Om','Guru','Raj','Sourav','Raj','Chinmaya','Om'],
    'Age':[20,21,22,22,22,23,20],
    'City':['Jsr','Cut','Ang','Bala','Ang','Pur','Jsr'],
}

df = pd.DataFrame(d)
df[df.duplicated()]

```

```python
df_last = df.drop_duplicates(keep='last')
df_last
```

```python
df_none = df.drop_duplicates(keep=False)
df_none
```
