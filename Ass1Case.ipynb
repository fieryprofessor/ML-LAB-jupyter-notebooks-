```python
#Q1
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
students = pd.read_csv("students.csv")

#(a) Average marks per student and top 5 performers
students['Average'] = students[['Math', 'Science', 'English']].mean(axis=1)
top5 = students.nlargest(5, 'Average')[['Name', 'Average']]
print("Top 5 performers:\n", top5, "\n")

#(b) Subject-wise average and standard deviation
subject_stats = students[['Math', 'Science', 'English']].agg(['mean', 'std']).T
print("Subject-wise stats:\n", subject_stats, "\n")

highest_var = subject_stats['std'].idxmax()
print(f"Subject with highest variation: {highest_var}\n")

#(c) Boxplot for each subject
students[['Math', 'Science', 'English']].boxplot()
plt.title("Distribution of Student Marks by Subject")
plt.ylabel("Marks")
plt.show()
```

```python
#Q2
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
sales = pd.read_csv("sales.csv")

#(a) Add Total and compute overall revenue
sales['Total'] = sales['Quantity'] * sales['Price']
overall_revenue = sales['Total'].sum()
print("Overall Revenue:", overall_revenue, "\n")

#(b) Group by region and find top region
rev_by_region = sales.groupby('Region')['Total'].sum().sort_values(ascending=False)
print("Revenue by region:\n", rev_by_region, "\n")

top_region = rev_by_region.idxmax()
print(f"Region contributing the most sales: {top_region}\n")

#(c) Histogram / bar chart of total sales per product
product_totals = sales.groupby('Product')['Total'].sum().sort_values(ascending=False)
product_totals.plot(kind='bar', figsize=(8,4))
plt.title("Total Sales per Product")
plt.xlabel("Product")
plt.ylabel("Total Revenue")
plt.tight_layout()
plt.show()

# Identify underperforming products (lowest 3)
underperformers = product_totals.nsmallest(3)
print("Underperforming Products:\n", underperformers)
```

```python
#Q3
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
auto_mpg = pd.read_csv("auto_mpg.csv")

#(a) Correlation of mpg with numeric features
corr = auto_mpg.corr(numeric_only=True)['mpg'].sort_values(ascending=False)
print("Correlation of MPG with other features:\n", corr, "\n")

# Identify key influencers
print("Features most influencing fuel efficiency:\n")
print("Strong negative correlation  lower mpg (e.g., heavy cars, high displacement)")
print("Strong positive correlation  higher mpg (e.g., newer models, lower cylinders)\n")

#(b) Scatter plot: weight vs mpg
plt.scatter(auto_mpg['weight'], auto_mpg['mpg'], alpha=0.6)
plt.title("Weight vs MPG")
plt.xlabel("Weight")
plt.ylabel("MPG")
plt.show()

# Observation
print("Trend: As weight increases, MPG decreases  heavier cars consume more fuel.")
```
