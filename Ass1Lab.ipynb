```python
#Q1
import numpy as np

#(a)
arr = np.arange(1, 21)

#(b) 
def is_prime(n):
    if n < 2: return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0: return False
    return True

primes = np.array([x for x in arr if is_prime(x)])

#(c)
mean_primes = primes.mean()
var_primes = primes.var(ddof=0) 
print("Primes:", primes)
print("Mean:", mean_primes)
print("Variance:", var_primes)
```

```python
#Q2
import numpy as np

#(a)
A = np.arange(1,17).reshape(4,4)

#(b) 
sub = A[2:4, 0:2]

#(c) 
det_sub = np.linalg.det(sub)
print("A=\n", A)
print("Sub=\n", sub)
print("Determinant of sub:", det_sub)
```

```python
#Q3
import pandas as pd

#(a)
df = pd.DataFrame({
    'Name': ['A', 'B', 'C', 'D', 'E'],
    'Math': [78, 85, 62, 90, 74],
    'Science': [82, 79, 88, 91, 70],
    'English': [75, 88, 65, 85, 80]
})

#(b)
df['Total'] = df[['Math','Science','English']].sum(axis=1)
df['Average'] = df['Total'] / 3

#(c)
top_idx = df['Average'].idxmax()
topper = df.loc[top_idx, ['Name','Average']]
print(df)
print("Topper:", topper['Name'], "with average", topper['Average'])
```

```python
#Q4
import numpy as np

#(a)
n = 1000
tosses = np.random.choice([0,1], size=n)  # 1=head, 0=tail

#(b)
heads = (tosses == 1).sum()
tails = (tosses == 0).sum()

#(c)
p_head = heads / n
print("Heads:", heads, "Tails:", tails)
print("Estimated P(head):", p_head, "Close to 0.5? ", abs(p_head - 0.5) < 0.05)
```

```python
#Q5
import pandas as pd

#(a)
df = pd.DataFrame({
    'ID':[1,2,3,4,5],
    'Name':['A','B','C','D','E'],
    'Salary':[40000, 60000, 35000, 50000, 45000]
})

#(b)
df['Bonus'] = 0.10 * df['Salary']

#(c)
avg_salary = df['Salary'].mean()
above_avg = df[df['Salary'] > avg_salary]
print(df)
print("Above average salary:\n", above_avg)
```

```python
#Q6
import numpy as np

#(a)
A = np.arange(1,10).reshape(3,3)
A = np.array([[1,2,3],[0,1,4],[5,6,0]])

#(b)
At = A.T
Ainv = np.linalg.inv(A)

#(c)
identity_approx = A.dot(Ainv)
print("A:\n", A)
print("Transpose:\n", At)
print("Inverse:\n", Ainv)
print("A * Ainv ≈\n", identity_approx)
```

```python
#Q7
import numpy as np
np.random.seed(0)

#(a)
temps = np.random.uniform(20, 40, size=30)  

#(b)
hottest_idx = temps.argmax()
coldest_idx = temps.argmin()
hottest = temps[hottest_idx]
coldest = temps[coldest_idx]

#(c)
mean_t = temps.mean()
median_t = np.median(temps)
std_t = temps.std(ddof=0)

print("Hottest day index:", hottest_idx, "Temp:", hottest)
print("Coldest day index:", coldest_idx, "Temp:", coldest)
print("Mean:", mean_t, "Median:", median_t, "Std:", std_t)
```

```python
#Q8
import pandas as pd

#(a)
s = pd.Series([55, 32, 48, 39, 90, 61, 28, 73])

#(b) 
s_replaced = s.astype(object)
s_replaced[s_replaced < 40] = 'Fail'

#(c) 
passed = (s >= 40).sum()

print(s_replaced)
print("Number passed:", passed)
```

```python
#Q9
import numpy as np
from collections import Counter

#(a)
rolls = np.random.randint(1,7, size=500)

#(b)
counts = Counter(rolls)

#(c)
relative_freq = {face: counts[face]/500 for face in range(1,7)}

print("Counts:", counts)
print("Relative frequencies:", relative_freq)
```

```python
#Q11
import pandas as pd
import numpy as np

#(a)
df = pd.DataFrame({
    'A':[1, np.nan, 3, np.nan],
    'B':[5, 6, np.nan, np.nan],
    'C':[9, 10, 11, 12]
})

#(b)
df_filled = df.fillna(df.mean())

#(c)
df_dropped = df[df.isna().sum(axis=1) <= 1]

print("Original:\n", df)
print("Filled:\n", df_filled)
print("Dropped (>1 missing removed):\n", df_dropped)
```

```python
#Q13
import pandas as pd

#(a)
df = pd.DataFrame({
    'Name':['A','B','C','D','E','F'],
    'Age':[19,21,18,22,20,17],
    'Marks':[65,72,55,81,68,93]
})

#(b)
overall_avg = df['Marks'].mean()
above_avg = df[df['Marks'] > overall_avg]

#(c)
young_high = df[(df['Age'] < 20) & (df['Marks'] > 60)]

print("Above overall average:\n", above_avg)
print("Names <20 and marks>60:\n", young_high['Name'].tolist())
```
