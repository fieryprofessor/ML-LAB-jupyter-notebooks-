```python
import pandas as pd
#Data Dictionary
marks= pd.DataFrame({
    'Om':[87,89,93,87,92],
    'Eva':[95,99,87,34,23],
    'Sam':[86,56,26,84,27],
    'Guru':[43,75,23,97,26],
    'Raj':[3,79,23,79,45]
},index=['a','b','c','d','e'])

#Printing
print(marks,"\n")
print(marks['Eva']['c'])
print(marks.Eva.a)
print(marks.at['d','Raj'])
print(marks.iat[0,1],"\n")

#Displaying O grade students
print("O grade:")
print(marks[marks>=90],"\n")

#Displaying A grade students
print("A grade:")
print(marks[(marks>=80) & (marks<90)])
```

```python
print("Transpose")
print(marks.T,"\n")

print("All marks sorted in Row indices:")
print(marks.sort_index(ascending=False),"\n") #Descending Order

print("All marks sorted in Column indices:")
print(marks.sort_index(axis=1)) 
```

```python
#Sort by values of the columns of a and e
print(marks.sort_values(by='a',axis=1,ascending=True),"\n")
print(marks.sort_values(by='e',axis=1,ascending=False),"\n")

#Sort by values of the rows of Om
print(marks.sort_values(by='Om',ascending=True),"\n")

#Select d row and sort values
print(marks.loc['d'].sort_values(ascending=True))
```

```python
import pandas as pd

# Define column names from UCI Auto MPG dataset
column_names = [
    "mpg", "cylinders", "displacement", "horsepower", "weight",
    "acceleration", "model_year", "origin", "car_name"
]

# URL of Auto MPG dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data"
url2="https://raw.githubusercontent.com/CourseMaterial/DataWrangling/main/flowerdataset.csv"

# Load dataset
df = pd.read_csv(
    url2,
    sep='\s+',   # data separated by spaces
    names=column_names,      # assign column names
    na_values="?"            # missing values marked as '?'
)

# Save to local CSV file
df.to_csv("auto_mpg.csv", index=False)

print("✅ Auto MPG dataset saved as 'auto_mpg.csv'")
print(df.head())
```
