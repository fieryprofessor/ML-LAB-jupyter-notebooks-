```python
#Wap to take a string as input and print its reverse without slicing
s = input()
for i in range(len(s)-1,-1,-1):
    print(s[i],end="")
```

```python
#Input a string and count how many vowels are present.
s = input()
c =0
for i in s:
    if (i=='a'or i=='e' or i=='i'or i=='o' or i=='u'||i=='A'or i=='E' or i=='I'or i=='O' or i=='U'):
        c +=1
print(c)
```

```python
#Input a sentence and return a dictionary with each word and its frequeny
s=input()
ns = s.split()
d={}
for i in ns:
    if i not in d:
        d[i] = 1
    else:
        d[i] += 1
print(d)
```

```python
#For a given 2D array, find sum of each row and each column separately.
import numpy as np
arr = np.array([[1,2,3],[4,5,6]])
arr_row1 = sum(arr[0])
arr_row2 = sum(arr[1])
arr_col1 = sum(arr[:,0])
arr_col2 = sum(arr[:,1])
arr_col3 = sum(arr[:,2])
print(arr_row1)
print(arr_row2)
print(arr_col1)
print(arr_col2)
print(arr_col3)
```
